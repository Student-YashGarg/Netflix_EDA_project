# Categorising [Vote_Average] Column
we would made 4 Categories: [Popular, average, below_avg, not_Popular] .

```python
def catigorize_col(df,col,labels):
    edges=[
        df[col].describe()['min'],
        df[col].describe()['25%'],
        df[col].describe()['50%'],
        df[col].describe()['75%'],
        df[col].describe()['max'] ]
    
    df[col]=pd.cut(df[col], edges, labels=labels, duplicates= 'drop')
    return df
```

```python
# # callling function
labels=['Not popular','below_avg','average','Popular']
catigorize_col(df,'Vote_Average',labels)
```

# Data Visualization

```python
sns.set_style('dark')
```

# 1.Which is the most frequent 'Genre' of movies released on Netflix ?

```python
g=sns.catplot(y='Genre' ,data=df ,kind='count' ,order=df['Genre'].value_counts().index 
              ,color='red' ,height=6)

# Access the actual Axes inside FacetGrid
ax = g.axes[0,0]

for container in ax.containers:
    ax.bar_label(container)

plt.title('Genre Column Distribution',fontsize=20,color='red',fontweight='bold')
plt.xlabel('Count',fontsize=18,color='black')
plt.ylabel('Genre',fontsize=18,color='black')
plt.show()
```

# 'Drama' is the most Frequent Genre in our Dataset among 19 other genres

# 2. Which has Highest votes in 'Vote_avg' column ?

```python
g=sns.catplot(y='Vote_Average' ,data=df ,kind='count' ,order=df['Vote_Average'].value_counts().index ,color='red')

ax=g.axes[0,0]

for container in ax.containers:
    ax.bar_label(container)
plt.title('Votes Distribution',fontsize=20,color='red',fontweight='bold')  
plt.show()
```

# 3. Average Genre Popularity

```python
# Sort genres by average popularity
genre_order = df.groupby('Genre',observed=True)['Popularity'].mean().sort_values(ascending=False).index
```

```python
ax = sns.barplot(
    y='Genre',
    x='Popularity',
    data=df,
    order=genre_order,
    color='red'
)

# Add value labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f')  # one decimal
    
plt.title('Average Genre Popularity',fontsize=20,color='red',fontweight='bold')  
plt.show()
```

# 4. Which movie got the Heighest Popularity and what's its Genre ?

```python
df[df['Popularity']==df['Popularity'].max()]
```

df['Popularity'].max() â†’ finds the highest popularity value

df[df['Popularity'] == ...] â†’ filters the row(s) that have that value

Returns a DataFrame, so if multiple movies share the same max popularity, all will be shown
